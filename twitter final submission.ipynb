```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
import re
```

```python
train_data = pd.read_csv('nlp-getting-started/train.csv')
```

```python
train_data.drop(['id','keyword','location'], axis =1, inplace =True)
```

```python
train_data
```

```python
ps = PorterStemmer()
```

```python
# lemmanitaziation
from nltk.stem import WordNetLemmatizer
wordnet = WordNetLemmatizer()
```

```python
def DataCleaning(dataSheet):
    corpus = []
    for i in range(0, len(dataSheet)):
        tweet = re.sub('[^a-zA-Z]',' ', dataSheet[i])
        tweet = tweet.lower()
#         tweet =  [ps.stem(word) for word in tweet.split() if word not in set(stopwords.words('english'))]
        tweet =  [wordnet.lemmatize(word) for word in tweet.split() if word not in set(stopwords.words('english'))]
        
        corpus.append(' '.join(tweet))
    return corpus    
        
```

```python
X = DataCleaning(train_data['text'])
```

```python
X 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
```

```python
# bagOfWords = cv.fit_transform(X).toarray()
tfIdf = tf.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(bagOfWords, train_data['target'], test_size =0.2, random_state =55, shuffle = True)
X_train, X_test, y_train, y_test = train_test_split(tfIdf, train_data['target'], test_size =0.2, random_state =55, shuffle = True)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred =model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
accuracy_score(y_pred,y_test)
f1_score(y_pred,y_test)
```
